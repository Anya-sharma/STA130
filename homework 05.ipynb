### Question 1:

1. **Key Factor for Testable Ideas**: The key factor is whether the idea is **quantifiable** and can be tested using **empirical data**. A hypothesis must be measurable in a way that allows for statistical evaluation, rather than relying on subjective or abstract concepts.

2. **Criteria for a Good Null Hypothesis**: A good null hypothesis is **clear, specific**, and **testable**. It should assert that there is **no effect** or **no difference**, making it possible to challenge or refute with data.

3. **Null Hypothesis vs. Alternative Hypothesis**: The **null hypothesis** (H₀) assumes no effect or no difference in the population being studied, acting as the default claim to test. The **alternative hypothesis** (H₁) proposes that there is an effect or difference, and the goal of hypothesis testing is to determine whether data provide enough evidence to reject the null hypothesis in favor of the alternative.

### Question 2:
### Relating Statistical Tests to Population Parameters

In hypothesis testing, we use the sample data points (xᵢ) to calculate the sample mean (x̄) in order to make inferences about the population mean (μ), and we compare this to the hypothesized population mean (μ₀) to determine whether to reject the null hypothesis.

### Question 3:
We imagine a world where the null hypothesis is true when calculating a p-value because the p-value tells us how likely it is to get our observed data (or something more extreme) if the null hypothesis were correct. By assuming the null hypothesis is true, we can measure how surprising or unlikely our sample results are in that scenario, which helps us decide whether to reject the null hypothesis.

### Question 4:
### Smaller p-values and the Null Hypothesis

A smaller p-value indicates that the observed data are so unusual that they would be highly unlikely to occur by random chance alone if the null hypothesis were true. This makes the null hypothesis seem increasingly "ridiculous" as the p-value decreases, strengthening the case for rejecting it in favor of the alternative hypothesis (H₁).

### Question 6:
### Proving or Disproving Hypotheses with p-values

**No, a smaller p-value cannot definitively prove that the null hypothesis is false**. It only provides evidence against the null hypothesis. Similarly, **it is not possible to definitively prove Fido's innocence or guilt using a p-value** (as discussed in the second pre-lecture video).

**P-values measure the probability of observing the obtained data, or more extreme data, assuming the null hypothesis is true.** A small p-value suggests that the observed data is unlikely if the null hypothesis were true, leading us to question its validity. However, it doesn't provide absolute proof.

Think of a courtroom analogy:

*   The null hypothesis is the presumption of innocence.
*   The p-value represents the evidence presented against the defendant.
*   A small p-value suggests strong evidence against the defendant, but it doesn't automatically guarantee a guilty verdict. Other factors, such as the quality of evidence and legal standards, also play a role.

**There is no specific p-value threshold that definitively proves or disproves a hypothesis**. Commonly used thresholds, like 0.05, are arbitrary conventions. Even with a very small p-value, there's always a small chance that the observed results occurred due to random variation, even if the null hypothesis is true.

**In conclusion, p-values provide a measure of evidence against the null hypothesis, but they don't offer definitive proof.** We can only use them to make informed decisions about whether to reject the null hypothesis based on the strength of the evidence and the context of the research question.


### Question 7:
Key Differences Between One-Sided and Two-Sided Tests
Two-Sided Test: A two-sided test checks for deviations in either direction from the null hypothesis. We test whether the effect is either significantly higher or lower than expected (i.e., testing for differences in both directions).
One-Sided Test: A one-sided test focuses on deviations in one direction only—either we test for an increase or a decrease, but not both.

```python
import pandas as pd  # Import libraries
import numpy as np

# Sample data loaded from personal GitHub
url = "https://raw.githubusercontent.com/Anya-sharma/STA130/refs/heads/main/DATA.csv"
df = pd.read_csv(url)  # Read the URL

# Set the random seed
np.random.seed(42)

# Set the number of bootstrap samples
bootstrap_samples = 1000

# Initialize arrays to store bootstrapped statistics
initial_mean_bootstrapped = np.zeros(bootstrap_samples)
final_mean_bootstrapped = np.zeros(bootstrap_samples)

# Bootstrapping process for InitialHealthScore
for i in range(bootstrap_samples):
    simulated_sample_initial = df['InitialHealthScore'].sample(n=len(df), replace=True)
    initial_mean_bootstrapped[i] = simulated_sample_initial.mean()

# Bootstrapping process for FinalHealthScore
for i in range(bootstrap_samples):
    simulated_sample_final = df['FinalHealthScore'].sample(n=len(df), replace=True)
    final_mean_bootstrapped[i] = simulated_sample_final.mean()

# Calculate the observed difference in means
observed_difference = df['FinalHealthScore'].mean() - df['InitialHealthScore'].mean()

# Calculate the differences of bootstrapped means
bootstrapped_differences = final_mean_bootstrapped - initial_mean_bootstrapped

# Calculate one-sided p-value (testing if FinalHealthScore > InitialHealthScore)
# Count how many bootstrapped differences are less than or equal to the observed difference
p_value_one_sided = np.mean(bootstrapped_differences <= observed_difference)

# Calculate 95% confidence intervals (optional)
initial_mean_ci = np.quantile(initial_mean_bootstrapped, [0.025, 0.975])
final_mean_ci = np.quantile(final_mean_bootstrapped, [0.025, 0.975])

# Print out the statistics to show the bootstrapped means and p-value
print("Bootstrapped 95% Confidence Intervals of Means:")
print("Initial Mean CI:", initial_mean_ci)
print("Final Mean CI:", final_mean_ci)
print(f"Observed Difference: {observed_difference}")
print(f"One-Sided P-Value: {p_value_one_sided}")
```

Key Changes Made:
Observed Difference Calculation: We compute the observed difference between the means of the FinalHealthScore and InitialHealthScore.
Bootstrapped Differences Calculation: We calculate the differences between bootstrapped means.
One-Sided P-value Calculation: The p-value is computed as the proportion of bootstrapped differences that are less than or equal to the observed difference, which corresponds to the hypothesis that the FinalHealthScore is greater than the InitialHealthScore.
Interpretation of the Hypothesis Test:
The one-sided test will assess if there is significant evidence to conclude that the FinalHealthScore is greater than the InitialHealthScore.
A smaller p-value will indicate stronger evidence against the null hypothesis (H₀) in favor of the alternative hypothesis (H₁).
Expectation for the P-value:
Since a one-sided test focuses on one direction, we might expect the p-value to be smaller compared to a two-sided test, where the probability is split between both tails of the distribution.

### Problem Introduction

The Fisher's Tea Experiment is a foundational example of hypothesis testing, designed by statistician Ronald Fisher to assess whether Dr. Muriel Bristol could distinguish between cups of tea where milk was poured first and those where tea was poured first. In a modern adaptation of this experiment, we test 80 STA130 students, with 49 of them correctly identifying the sequence in which tea or milk was poured. 

The goal of this analysis is to determine whether the STA130 students' results could be explained by random guessing (chance), or if the students demonstrate a genuine ability to identify the pouring sequence beyond what would be expected from guessing alone.

### Relationship Between This Experiment and the Original Fisher's Experiment

While the original experiment centered on Dr. Bristol's claim of distinguishing tea preparation methods, this modern version tests a group of students who likely lack expertise in tea tasting. The original experiment involved only 8 trials, with the parameter of interest being a personalized ability specific to Bristol. In contrast, the STA130 experiment involves a larger group of 80 students, making the focus less about individual expertise and more about general sensory perception across a larger, less specialized population.

### Null and Alternative Hypotheses

**Null Hypothesis (H₀)**:  
The STA130 students are guessing randomly when identifying whether tea or milk was poured first. The probability of correctly identifying the sequence is 50% (i.e., \( p = 0.5 \)).

**Informal Interpretation of H₀**:  
The students have no real ability to detect whether tea or milk was poured first; their answers are purely based on chance.

**Alternative Hypothesis (H₁)**:  
The STA130 students can correctly distinguish the pouring sequence at a rate higher than 50%, suggesting that their ability to identify the sequence is greater than what would be expected by chance alone. \( p > 0.5 \).

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom_test

# Reproducibility
np.random.seed(130)

# Given data
correct_answers = 49
total_students = 80
probability_guessing = 0.5  # Under the null hypothesis, students are guessing randomly

# 1. Histogram of correct answers
data = np.random.binomial(total_students, probability_guessing, 10000)  # Simulating random guessing outcomes
plt.hist(data, bins=15, color='skyblue', edgecolor='black')
plt.axvline(correct_answers, color='red', linestyle='dashed', linewidth=2)
plt.title('Histogram of Simulated Correct Answers (Under H0)')
plt.xlabel('Number of Correct Answers')
plt.ylabel('Frequency')
plt.show()
```

```python
import numpy as np

# Set seed for reproducibility
np.random.seed(130)

# Parameters
n_students = 80
p_null = 0.5  # Null hypothesis: random guessing
n_simulations = 10000  # Number of simulations
observed_statistic = 49 / 80  # 49 correct out of 80
population_parameter_value_under_H0 = 0.5  # Expected proportion under H0

# Simulate the sampling distribution under H0
simulated_results = np.random.binomial(n_students, p_null, n_simulations)
simulated_statistics = simulated_results / n_students

# "As or more extreme" (symmetric) calculation
SimStats_as_or_more_extreme_than_ObsStat = \
    abs(simulated_statistics - population_parameter_value_under_H0) >= \
    abs(observed_statistic - population_parameter_value_under_H0)

# p-value calculation
p_value = SimStats_as_or_more_extreme_than_ObsStat.sum() / n_simulations

# Output p-value
p_value
```
