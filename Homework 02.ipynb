````python
#Part 2:
import numpy as np

# Initialize variables
all_doors = [1, 2, 3]
initial_choice = 1
wins = 0
reps = 100000

for _ in range(reps):
    # Randomly select the winning door
    winning_door = np.random.choice(all_doors)

    # Host reveals a goat door (not the player's choice or the winning door)
    remaining_doors = [door for door in all_doors if door != initial_choice and door != winning_door]
    revealed_goat = np.random.choice(remaining_doors)

    # Swap to the remaining door
    switched_choice = [door for door in all_doors if door != initial_choice and door != revealed_goat][0]

    # Check if the switched choice is the winning door
    if switched_choice == winning_door:
        wins += 1

# Calculate and print the win percentage
win_rate = wins / reps
print(win_rate)
````
